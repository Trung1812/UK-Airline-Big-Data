```python
import pandas as pd
import json

# Load the JSON file
with open("test.json", "r") as f:
    json_data = json.load(f)

# Extract the 'data' section
data = json_data["data"]

# Flatten the nested structure into a DataFrame
df = pd.json_normalize(data)

# Display the DataFrame
print(df)
```

```python
# Count flights where both 'departure.actual' and 'arrival.actual' are not null
landed_flights_count = df[df['flight_status']=='landed'].shape[0]

print(f"Number of flights that have landed: {landed_flights_count}")
```
